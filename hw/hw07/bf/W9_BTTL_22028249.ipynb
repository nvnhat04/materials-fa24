```python
# Run this cell to set up the notebook, but please don't change it.

# These lines import the Numpy and Datascience modules.
import numpy as np
from datascience import *

# These lines do some fancy plotting magic.
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
data = Table.read_table('revenue_data.csv')
data.show(5)
```

```python
profit_before = data['Revenue_Before']
profit_after = data['Revenue_After']
```

```python
data.hist(bins = np.arange(0, 15, 1))
```

```python
import math
import matplotlib.pyplot as plt


# Tnh hiu gia cc gi tr
differences = [profit_before[i] - profit_after[i] for i in range(len(profit_before))]

# Trung bnh ca hiu
mean_d = sum(differences) / len(differences)

#  lch chun ca hiu
variance_d = sum((d - mean_d) ** 2 for d in differences) / (len(differences) - 1)
std_d = math.sqrt(variance_d)

# Tnh gi tr t
t_stat = mean_d / (std_d / math.sqrt(len(differences)))

# V biu  so snh doanh thu trc v sau
plt.figure(figsize=(10, 5))
plt.plot(profit_before, label="Trc khi c web mi", marker="o", linestyle="--", alpha=0.7)
plt.plot(profit_after, label="Sau khi c web mi", marker="s", linestyle="-", alpha=0.7)

# nh dng biu 
plt.xlabel("Ngy bn hng")
plt.ylabel("Doanh thu")
plt.title("So snh doanh thu trc v sau khi trin khai web mi")
plt.legend()
plt.grid(True)
plt.show()

print("T-statistic:", t_stat)
```
